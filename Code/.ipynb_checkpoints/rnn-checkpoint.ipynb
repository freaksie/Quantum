```python
x_train=x_train.reshape((x_train.shape[0],int(x_train.shape[1]/10),10))
x_test=x_test.reshape((x_test.shape[0],int(x_test.shape[1]/10),10))
print(x_train.shape,x_test.shape)
```

```python
def arch():
    model=Sequential()
    model.add(Input(shape=(None,10)))
    model.add(layers.GRU(units=32,return_sequences=True,activation='relu',kernel_regularizer=L2(0.001)))
    model.add(layers.GRU(units=32,return_sequences=False,activation='relu',kernel_regularizer=L2(0.001)))
    model.add(layers.Dense(16,activation='relu'))
    model.add(layers.Dense(8,activation='relu'))
    model.add(layers.Dense(1,activation='sigmoid'))
    return model
# arch2().summary()
```

```python
model=arch()
model.summary()
# opt=SGD( learning_rate=0.001, momentum=0.9)
custom_optimizer = Adam(learning_rate=0.001, use_ema=True, ema_momentum=0.9)
model.compile(loss=binary_crossentropy, optimizer=custom_optimizer, metrics=['accuracy'])
early_stopping = EarlyStopping(monitor='val_loss', patience=9, restore_best_weights=True, min_delta=0.002)
```

```python
with tf.device("/gpu:0"):
    H=model.fit(x_train,y_train,
            validation_data=(x_test, y_test),
            epochs=100,batch_size=64,
            callbacks=[early_stopping],
            verbose=1)
```

```python
N = 100
plt.style.use("ggplot")
plt.figure()
plt.plot(np.arange(0, N), H.history["loss"], label="train_loss")
plt.plot(np.arange(0, N), H.history["val_loss"], label="test_loss")
plt.title("GRU")
plt.xlabel("Epoch #")
plt.ylabel("loss")
plt.legend(loc="lower left")
```
